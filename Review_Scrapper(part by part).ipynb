```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
searchString = 'Iphone7'
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + searchString
```

```python
flipkart_url
```

```python
#Ureq - Alias for urlopen and its responsible for to open any url
#Ureq is the part of the package called urllib
uClient = uReq(flipkart_url)
uClient
#here uClient gives some response - which simply means it open the url and initiate the search operation
```

```python
#open the link and do the read operation
flipkartPage = uClient.read()
flipkartPage
#here i got the entire html page of the website or search page
```

```python
#close the connection after the reading operation has done
uClient.close()
```

```python
#Beautifulsoup - It tries to remove the unnecessary text or information and it create some date in the form of structured format
#Not fully structured format, somehow structured format
#bs4 will works fine for html parser, xml parser
flipkart_html = bs(flipkartPage,'html.parser')
flipkart_html
```

```python
#Try to filter out the class that containing the review, so we need to click the url of each and every product that listed out on the page
#inside the div -- > class --> we  have the link(href)
bigboxes = flipkart_html.findAll("div",{"class":"_1AtVbE col-12-12"})
```

```python
bigboxes
```

```python
#I don't want the some of the div classes so i removed it.
del bigboxes[0:3]
```

```python
#I extract the data for the first(iphone product) div box here.
box = bigboxes[0]
box
```

```python
len(bigboxes)
#We have 27 boxes(products we have totally)
#so we need to hit all the 27 boxes to extract the inforamtion
```

```python
box.div
```

```python
#inside the multiple div's i extracted the href link
box.div.div.div.a['href']
```

```python
#Append the extracted url with my base url
productLink="https://www.flipkart.com" + box.div.div.div.a['href']
productLink
#If you copy the below link and try to open in the browser then you can able to land on the product page
```

```python
#once i reached the product link now i want to extract the review of the product 
prodRes = requests.get(productLink)
prodRes.encoding = 'utf-8'
```

```python
prodRes
#200 - good response 
```

```python
prod_html = bs(prodRes.text,'html.parser')
print(prod_html)
```
